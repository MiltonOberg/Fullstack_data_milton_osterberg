```python
df['discounted_price']=df['discounted_price'].str[1:].str.replace(",","").astype(float)
```

```python
df.head()
```

```python
df=df[df['rating_count'].isnull()==False]
```

```python
df.info()
```

```python
df['actual_price']=df['actual_price'].str[1:].str.replace(",","").astype(float)
```

```python
df.head()
```

```python
df['rating_count']=df['rating_count'].str.replace(",","").astype(float)
```

```python
df.head()
```

```python
# median rating count osv

if 'rating_count' in df.columns:
    
    mean_rating = df['rating_count'].mean()
    median_rating = df['rating_count'].median()
    min_rating = df['rating_count'].min()
    max_rating = df['rating_count'].max()
    percentile_10 = df['rating_count'].quantile(0.10)
    percentile_90 = df['rating_count'].quantile(0.90)

    print("Rating Statistics:")
    print(f"Mean Rating: {mean_rating:.2f}")
    print(f"Median Rating: {median_rating:.2f}")
    print(f"Min Rating: {min_rating:.2f}")
    print(f"Max Rating: {max_rating:.2f}")
    print(f"10th Percentile: {percentile_10:.2f}")
    print(f"90th Percentile: {percentile_90:.2f}")
else:
    print("Column 'rating_count' does not exist in the dataset.")
```

```python
df["category"]= df["category"].str.split("|").str[0]
df["category"].head()
```

```python
unique_categories= df["category"].unique()
unique_categories
```

```python
mean_prices= df.groupby("category")["actual_price"].mean().reset_index()
```

```python
mean_prices.sort_values("actual_price", ascending= False, inplace= True)
```

```python
mean_prices
```

```python
import matplotlib.pyplot as plt
mean_prices.head(10).plot(kind="bar", x="category", y="actual_price", title="Top 10 Most Expensive Products by Category")
```
